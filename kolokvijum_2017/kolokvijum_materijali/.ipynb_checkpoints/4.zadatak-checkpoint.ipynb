Na osnovu uzorkovanja signala oblika $x(t)=cos(2*\pi*f*t)$ u trajanju od 0.5 sekundi dobijene su 32 vrednosti čiji je grafik prikazan u gornjem delu slike, a diskretnom Furijeovom transformacijom signala dobijena je spektralna analiza prikazana na donjem delu slike. 
<img src='4.zadatak.png'>
* Kolika je frekvencija $f_s$ uzorkovanja signala? 
* Kolika je osnovna frekvencija $f$ signala? 
* Nacrtati frekvencijski spektar za signal čija je osnovna frekvencija $f+1$ i objasniti promene do kojih dolazi. 

```python
import numpy as np
Fs = 64
print("Sample frequency: {}Hz".format(Fs))
```

```python
print("Signal frequency: {}Hz".format(Fs/2))
```

```python
def signal(t, f):
    return np.cos(2*np.pi*f*t)
```

```python
t = np.linspace(0, 35, Fs)
f = signal(t, 64)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(t, f)
```
