### Diskretna Furijeova transformacija


<img src='fourier_transform.png', width=500>

```python
import numpy as np 
import matplotlib.pyplot as plt
# u numpy biblioteci postoji paket fft za rad sa Furijeovim transformacijama
from numpy import fft as f

%matplotlib inline
```

```python
# y je signal koji odgovara zvucnom signalu telefonskih centrala
t = np.linspace(0, 2*np.pi, 1024)
y = np.sin(2*np.pi*350*t)+np.sin(2*np.pi*440*t)
```

```python
# zbog velike frekvencije signala, suzavamo opseg prikaza na mali interval
plt.xlim(0, 0.8)
plt.plot(t, y)
```

```python
# funkcija fft izracunava koeficijente Furijeove transformacije
c = f.fft(y)
```

```python
# koeficijenti koji se dobijaju su kompleksni brojevi
type(c), c.shape, c.dtype
```

```python
# za koeficijente vazi uredjenje
# c[0] - zero-frequency clan 
# c[1], ...., c[n/2] - positive frequency clanovi 
# c[n/2+1], ...., c[n-1] - negative frequency clanovi 

c
```

```python
# funkcija koja prerasporedjuje koeficijente tako da prvo idu koeficijenti koji odgovaraju 
# negativnim frekvencijama, pa nulti koeficijent, pa koeficijenti koji odgovaraju pozitivnim frekvencijama
c = f.fftshift(c)
```

```python
# generisemo niz frekvencija sinusoida
freq = np.arange(-512, 512, 1)

# moduo kompleksnog broja odgovara amplitudi sinusoide
amp = np.abs(c)

# ugao odgovara faznom pomeraju sinusoide
pha = np.angle(c)
```

```python
plt.plot(freq, amp)
```

```python
plt.plot(freq, pha)
```

```python
# praksa je da se za analizu frekvencijskog spektra generisu tzv. stem grafikoni
# npr. prikaz srednjih 20 frekvencija
freq_stem=np.arange(-20, 21, 1)
amp_stem=amp[492:533]
plt.stem(freq_stem, amp_stem)
```
